```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
perfil = pd.read_csv('../data/clear_perfil_visitor_and_students.csv')
```

```python
def extract_numerical_columns(df):
    """
    Extracts numerical columns from a DataFrame.

    Parameters:
    - df : Input DataFrame.

    Returns:
    - DataFrame containing only numerical columns.
    """
    numerical_df = df.select_dtypes(include=['int64', 'float64'])
    return numerical_df

numerical_df = extract_numerical_columns(perfil)

```
